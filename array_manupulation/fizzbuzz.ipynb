# Array Manupulation

### Question : Implement the FuzzBuzz problem.


The FizzBuzz problem is a classic programming problem often used as an exercise in coding interviews. 

In this solution:

We define a function fizzbuzz that takes an integer n as input.

    - We loop through the numbers from 1 to n.
    - For each number, we check if it is divisible by 3 and 5. If so, we print "FizzBuzz".
    - If it is only divisible by 3, we print "Fizz".
    - If it is only divisible by 5, we print "Buzz".
    - If it is not divisible by either 3 or 5, we print the number itself.
You can test the function with different ranges by passing different values to the fizzbuzz function.

Here's a Python solution for the FizzBuzz problem:

```python
def fizzbuzz(n):
    for i in range(1, n+1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Test the function with an example range (1 to 100)
fizzbuzz(100)
```
